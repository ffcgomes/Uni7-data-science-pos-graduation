```python
import wikipedia
wikipage = wikipedia.page('coronavirus').content
print(wikipage)
```

```python
import nltk
from nltk.tokenize import sent_tokenize, PunktSentenceTokenizer
nltk.download('punkt')
```

```python
wiki_sent = sent_tokenize(wikipage)
print(wiki_sent)
```

```python
from nltk.tokenize import word_tokenize
```

```python
wiki_words = []
for sent in wiki_sent:
    wiki_words.extend(word_tokenize(sent))
print(wiki_words)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stopwords = stopwords.words('english')
```

```python
word_frequencies = {}
for word in wiki_words:
    if(word not in stopwords and word.isalpha()):
        if(word not in word_frequencies.keys()):
            word_frequencies[word] = 1
        else:
            word_frequencies[word]+=1
print(word_frequencies)
        
```

```python
maximum_frequency = max(word_frequencies.values())
```

```python
maximum_frequency
```

```python
for word in word_frequencies.keys():
    word_frequencies[word] /=maximum_frequency
print(word_frequencies)
```

```python
max(word_frequencies.values())
```

```python
sentences_scores = {}
for sent in wiki_sent:
    for word in word_tokenize(sent):
        if(word in word_frequencies.keys()):
            if(sent not in sentences_scores.keys()):
                sentences_scores[sent] = word_frequencies[word]
            else:
                sentences_scores[sent] += word_frequencies[word]
for sent in sentences_scores.keys():
    print(sent, '--->', sentences_scores[sent])
```

```python
import heapq
summary_sentences = heapq.nlargest(10,sentences_scores, key=sentences_scores.get)
print(summary_sentences)
```
