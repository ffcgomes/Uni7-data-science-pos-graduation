```python
#x = df.loc[df['Ano'].isin(['2008','2009','2010'])]
df = df.loc[df['Ano'].isin(['2007'])==False]
print(df)
```

```python
df=df.drop(['Ano','Semana'], axis=1)
df
```

```python
#Normalizando os atributos 

df=(df-df.mean())/df.std() 

plt.xlabel('Data')
plt.ylabel('Precipita√ß√£o')

plt.plot(df['Infectados'])
plt.plot(df['Precipitacao'])

plt.gca().legend(('Infectados','Precipitacao'))

plt.show()
```
