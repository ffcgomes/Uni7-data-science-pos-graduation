```python
import pandas as pd
d=pd.read_csv("Sobral_Semanal_IBGE.csv", sep=",")
df = pd.DataFrame(data=d)
#print(len(df))
```

```python
def inserir_linha(idx, df, df_inserir):
    dfA = df.iloc[:idx, ]
    dfB = df.iloc[idx:, ]
    df = dfA.append(df_inserir).append(dfB).reset_index(drop = True)
    return df
```

```python
df.loc[0,'Populacao']
j=0
n=1
d_ano=0
while n<=676:
    i=1
    j=j+1
    #print(row['Populacao'])
    while i<7:
        d_inserido = d = {'Dia': [i+1],'Semanas':df.iloc[j-1,1],
        'Semana':df.iloc[j-1,2],'Ano':df.iloc[j-1,3],
        'Casos de Dengue':df.iloc[j-1,4],'IDH':df.iloc[j-1,5],
        'Populacao':df.iloc[j-1,6],'Populacao Sexo Feminino':df.iloc[j-1,7],
        'Populacao Sexo Masculino':df.iloc[j-1,8],'Idade< 15':df.iloc[j-1,9],
        'Idade-15 – 64':df.iloc[j-1,10],'Idade> 65':df.iloc[j-1,11],
        'Densidade Demográfica':df.iloc[j-1,12],'Populacao urbana':df.iloc[j-1,13],
        'Populacao Rural':df.iloc[j-1,14],'Mortalidade Infantil':df.iloc[j-1,15],
        'Infectados':df.iloc[j-1,16],}
        df_inserido = pd.DataFrame(data = d_inserido)
        df = inserir_linha(j, df, df_inserido)

        i=i+1
        j=j+1
    n=n+1
    #print(i,j,n,d_ano)
    
df.to_csv('Sobral_Diario_IBGE.csv')
```
